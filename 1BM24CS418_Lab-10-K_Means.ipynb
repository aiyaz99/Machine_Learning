To Do Implementation: K-means algorithm Clustering

```python
import pandas as pd
import numpy as np

# Load CSV file
data = pd.read_csv('iris (2).csv')

# Select only numeric columns
X = data.select_dtypes(include=[np.number]).values

# Number of clusters
k = 3

# Step 1: Initialize centroids randomly
np.random.seed(42)
centroids = X[np.random.choice(len(X), k, replace=False)]

def compute_distance(a, b):
    return np.linalg.norm(a - b)

for iteration in range(100):
    clusters = [[] for _ in range(k)]

    # Step 2: Assign points to nearest centroid
    for point in X:
        distances = [compute_distance(point, c) for c in centroids]
        cluster_index = np.argmin(distances)
        clusters[cluster_index].append(point)

    # Step 3: Update centroids
    new_centroids = []
    for cluster in clusters:
        if cluster:
            new_centroids.append(np.mean(cluster, axis=0))
        else:
            new_centroids.append(centroids[len(new_centroids)])

    new_centroids = np.array(new_centroids)

    # Step 4: Check convergence
    if np.allclose(centroids, new_centroids):
        break

    centroids = new_centroids

print("Final Centroids:\n", centroids)

# Assign final labels
labels = []
for point in X:
    distances = [compute_distance(point, c) for c in centroids]
    labels.append(np.argmin(distances))

print("Cluster Labels:\n", labels)

# ✅ ADD CLUSTER COLUMN TO ORIGINAL DATA
data['Cluster'] = labels

# ✅ EXPORT TO CSV FILE
output_file = 'kmeans_output.csv'
data.to_csv(output_file, index=False)

print(f"Output saved to {output_file}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv('iris (2).csv')

# Use only required features
X = data[['petal_length', 'petal_width']].values

# -----------------------------------
# ✅ Step 1: Scaling (Standardization)
# -----------------------------------
mean = np.mean(X, axis=0)
std = np.std(X, axis=0)
X_scaled = (X - mean) / std

# -----------------------------------
# ✅ Function for K-Means (Manual)
# -----------------------------------
def kmeans_manual(X, k, max_iter=100):
    np.random.seed(42)
    centroids = X[np.random.choice(len(X), k, replace=False)]

    for _ in range(max_iter):
        clusters = [[] for _ in range(k)]

        # Assign clusters
        for point in X:
            distances = [np.linalg.norm(point - c) for c in centroids]
            cluster_idx = np.argmin(distances)
            clusters[cluster_idx].append(point)

        # Update centroids
        new_centroids = []
        for cluster in clusters:
            if cluster:
                new_centroids.append(np.mean(cluster, axis=0))
            else:
                new_centroids.append(centroids[len(new_centroids)])

        new_centroids = np.array(new_centroids)

        # Check convergence
        if np.allclose(centroids, new_centroids):
            break

        centroids = new_centroids

    # Calculate WCSS (inertia)
    wcss = 0
    labels = []
    for point in X:
        distances = [np.linalg.norm(point - c) for c in centroids]
        cluster_idx = np.argmin(distances)
        labels.append(cluster_idx)
        wcss += distances[cluster_idx]**2

    return centroids, labels, wcss

# -----------------------------------
# ✅ Step 2: Elbow Method
# -----------------------------------
wcss_values = []

for k in range(1, 11):
    _, _, wcss = kmeans_manual(X_scaled, k)
    wcss_values.append(wcss)

# Plot elbow graph
plt.figure()
plt.plot(range(1, 11), wcss_values, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.title('Elbow Method')
plt.show()

# -----------------------------------
# ✅ Step 3: Final Model (k=3)
# -----------------------------------
k = 3
centroids, labels, _ = kmeans_manual(X_scaled, k)

print("Final Centroids:\n", centroids)

# -----------------------------------
# ✅ Step 4: Add Cluster & Export CSV
# -----------------------------------
data['Cluster'] = labels
data.to_csv('iris_clustered_manual.csv', index=False)

print("Output saved as iris_clustered_manual.csv")

# -----------------------------------
# ✅ Step 5: Visualization
# -----------------------------------
plt.figure()
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels)
plt.xlabel('Petal Length (scaled)')
plt.ylabel('Petal Width (scaled)')
plt.title('Manual K-Means Clustering')
plt.show()
```
